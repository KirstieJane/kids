```r

n_color_age_facet <- ggplot(data, aes(x=motion_thresh, y=med_rsq, group=(n), color=(n)))+
	geom_errorbar(aes(ymin=med_rsq-CI_95/2, 
			  ymax=med_rsq+CI_95/2), 
		      width=0, alpha=.6)+
	facet_wrap(~age_l)+
	coord_cartesian(y=c(.5, 1))+
	scale_color_gradient(low='gray', high='darkblue')

age_color_n_facet <- ggplot(data, aes(x=motion_thresh, y=med_rsq, group=factor(age_l), color=factor(age_l)))+
	geom_errorbar(aes(ymin=med_rsq-CI_95/2, 
			  ymax=med_rsq+CI_95/2), 
		      width=0, alpha=.2)+
	coord_cartesian(y=c(.8, 1))+
	facet_wrap(~n)

plotThing <- function(aPlot){
	aPlot + 
		geom_point(alpha=.2)+
		geom_line(alpha=.5)+
		scale_x_continuous(trans='identity')+
		scale_y_continuous(trans='identity')
}
```

```r

plotThing(n_color_age_facet)
```

```r
plotThing(age_color_n_facet)
```
