## Now plotting pearson R

```r
aplot <-ggplot(data, aes(x=motion_thresh, y=med_pear_r, group=n, color=n)) +scale_colour_gradientn(colours=rainbow(4))+
geom_errorbar(aes(ymin=med_pear_r-pear_r_CI/2, 
			  ymax=med_pear_r+pear_r_CI/2), 
		      width=0, alpha=.6)+geom_point()+
		geom_line(alpha=.5)


print(aplot)
#ggsave(aplot, filename='/Users/Jlnrd/Desktop/mean_by_perc_fd.jpg',  width=6, height=6)#, units='in', dpi=150) 
```

## Now plotting in spearman

```r
aplot <-ggplot(data, aes(x=motion_thresh, y=med_spear_r, group=n, color=n)) +scale_colour_gradientn(colours=rainbow(4))+
geom_errorbar(aes(ymin=med_spear_r-spear_R_CI/2, 
			  ymax=med_spear_r+spear_R_CI/2), 
		      width=0, alpha=.6)+geom_point()+
		geom_line(alpha=.5)


print(aplot)
```
